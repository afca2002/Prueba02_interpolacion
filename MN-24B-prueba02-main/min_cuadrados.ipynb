La ecuación de la línea es:

$ y(x) = a_1 x + a_0 $

Al realizar el proceso de mínimos cuadrados queda el siguiente sistema de ecuaciones:


$( \sum_i (y_i - a_1 x_i - a_0) , \sum_i (y_i - a_1 x_i - a_0)x_i ) = 0$

```python
# Derivadas parciales para regresión lineal
# ####################################################################
def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_1 * a_1 + c_0 * a_0 = c_ind


    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_1``: coeficiente del parámetro 1.

    ``c_0``: coeficiente del parámetro 0.

    ``c_ind``: coeficiente del término independiente.

    """

    # coeficiente del término independiente
    c_ind = sum(ys)

    # coeficiente del parámetro 1
    c_1 = sum(xs)

    # coeficiente del parámetro 0
    c_0 = len(xs)

    return (c_1, c_0, c_ind)


def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 0 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_1 * a_1 + c_0 * a_0 = c_ind


    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return

    ``c_1``: coeficiente del parámetro 1.

    ``c_0``: coeficiente del parámetro 0.

    ``c_ind``: coeficiente del término independiente.

    """
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        c_ind += xi * yi

        # coeficiente del parámetro 1
        c_1 += xi * xi

        # coeficiente del parámetro 0
        c_0 += xi

    return (c_1, c_0, c_ind)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
xs1 = np.array([-5.0000, -3.8889, -2.7778, -1.6667, -0.5556, 0.5556, 1.6667, 2.7778, 3.8889, 5.0000])
ys1 = np.array([57.2441, 33.0303, 16.4817, 7.0299, 0.5498, 0.7117, 3.4185, 12.1767, 24.9167, 44.2495])

# Construcción de matrices para mínimos cuadrados
X = np.vstack([xs1**2, xs1, np.ones(len(xs1))]).T
Y = ys1

# Resolviendo el sistema
coeffs = np.linalg.lstsq(X, Y, rcond=None)[0]
a2, a1, a0 = coeffs

# Generar la curva ajustada
x_fit = np.linspace(min(xs1), max(xs1), 100)
y_fit = a2 * x_fit**2 + a1 * x_fit + a0

# Graficar
plt.scatter(xs1, ys1, label="Datos")
plt.plot(x_fit, y_fit, color="red", label=r"$y = a_2x^2 + a_1x + a_0$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste cuadrático - Conjunto de datos 1")
plt.legend()
plt.show()
```

```python
# Datos
xs2 = np.array([0.0003, 0.0822, 0.2770, 0.4212, 0.4403, 0.5588, 0.5943, 0.6134, 0.9070, 1.0367, 1.1903, 1.2511, 1.2519, 1.2576, 1.6165, 1.6761, 2.0114, 2.0557, 2.1610, 2.6344])
ys2 = np.array([1.1017, 1.5021, 0.3844, 1.3251, 1.7206, 1.9453, 0.3894, 0.3328, 1.2887, 3.1239, 2.1778, 3.1078, 4.1856, 3.3640, 6.0330, 5.8088, 10.5890, 11.5865, 11.8221, 26.5077])

# Linealización
log_ys2 = np.log(ys2)

# Construcción de matrices para mínimos cuadrados
X_exp = np.vstack([xs2, np.ones(len(xs2))]).T
Y_exp = log_ys2

# Resolviendo el sistema
coeffs_exp = np.linalg.lstsq(X_exp, Y_exp, rcond=None)[0]
b, log_a = coeffs_exp
a = np.exp(log_a)

# Generar la curva ajustada
x_fit_exp = np.linspace(min(xs2), max(xs2), 100)
y_fit_exp = a * np.exp(b * x_fit_exp)

# Graficar
plt.scatter(xs2, ys2, label="Datos")
plt.plot(x_fit_exp, y_fit_exp, color="green", label=r"$y = a \cdot e^{bx}$")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste exponencial - Conjunto de datos 2")
plt.legend()
plt.show()
```
